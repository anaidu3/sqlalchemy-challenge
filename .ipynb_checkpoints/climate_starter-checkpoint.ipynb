# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect, asc, desc
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
# refer to "keys"
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
Session = Session(engine)
```

```python
Inspect = inspect(engine)
```

```python
Inspect.get_table_names()
```

```python
#column details for measurement
col = Inspect.get_columns('measurement')
for cols in col:
    print(cols["name"], cols["type"])
```

```python
#column details for station
col = Inspect.get_columns('station')
for cols in col:
    print(cols["name"], cols["type"])
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
# date_last = Session.execute("SELECT MAX(date) FROM Measurement").fetchall()
# print(date_last)
date_last = Session.query(func.max(Measurement.date)).all()
date_end = date_last[0][0]
date_end
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
Struct = dt.date.today()
date_last_clean = Struct.replace(year=int(date_end[:4]), month=int(date_end[5:7]), day=int(date_end[8:]))
date_last_year = date_last_clean - dt.timedelta(days=365)
# Calculate the date one year from the last date in data set.
date_start = date_last_year.strftime("%Y-%m-%d")
print(date_start)
```

```python
# Perform a query to retrieve the data and precipitation scores
# Sort the dataframe by date
prcp = Session.query(Measurement.date,Measurement.prcp)\
                          .filter(Measurement.date <= date_end)\
                          .filter(Measurement.date >= date_start)\
                          .order_by(Measurement.date.asc()).all()
prcp
# # Save the query results as a Pandas DataFrame and set the index to the date column
df = pd.DataFrame(prcp).rename(columns={0:"Date",1:"Precipitation Score"})

df.set_index("Date",inplace=True)
df.head()

# Use Pandas Plotting with Matplotlib to plot the data
df.plot(figsize=(9,4),ylim=(0,8),title=f'Precipitation Score', label ="Precipitation", rot=45)
plt.legend(loc="upper right")
plt.savefig("Resources/1_Precipitation_Score.jpg")
plt.show()

# Use Pandas to calculate the summary statistics for the precipitation data
df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
stations_total = Session.query(Station).count()
print(f'The total number of stations in the dataset is {stations_total}.')
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
stations_active = Session.query(Measurement.station,func.count(Measurement.station))\
                                .group_by(Measurement.station).order_by(desc(func.count(Measurement.station))).all()
station_highest = stations_active[0]
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
Session.query(func.min(Measurement.tobs),func.max(Measurement.tobs),func.avg(Measurement.tobs)).filter(Measurement.station == station_highest[0]).all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_data = Session.query(Measurement.date,Measurement.station,Measurement.tobs)\
             .filter(Measurement.date >= date_start).filter(Measurement.date<=date_end)\
             .filter(Measurement.station == station_highest[0]).all()       
```

```python
temp_df = pd.DataFrame(temp_data, columns=["Date","Station ID","Temperature_Recorded"])

temp_df.set_index("Date",inplace=True)
temp_df.head()

# Use Pandas Plotting with Matplotlib to plot the data
temp_df.plot.hist(by="Temperature_Recorded",bins=12,label="t",title ="Temperature for Station")
plt.ylim=(0,80)
plt.ylabel("Frequency of Temp")
plt.xlabel("Temperature (F)")
plt.legend(loc="upper right")
plt.savefig("Resources/2_Temperature.jpg")
plt.show()
```
